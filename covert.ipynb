```python
from pandas import read_csv
```

```python
df = read_csv("./data/dataset.csv")
df.columns
```

```python
df = df[['Title','FullDescription']]
df.head()
```

```python
text_col = []
for _,row in df.iterrows():
    prompt = f"Here is a job description for a {row['Title']}. I want you to learn this job description and be able to generate similar job descriptions that are also considered to be best for the given job title\n\ndescription : \n{row['FullDescription']}\n\n"
    print(len(row['FullDescription']))
    text_col.append(prompt)
```
